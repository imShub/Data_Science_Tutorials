# Basic Charting using Matplotlib

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([1,2,4,5],[1,2,4,5])
plt.show()
```

## Use of Titles and Lables

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([1,2,4,5],[1,2,4,5])
plt.title("Example of Title")
plt.xlabel("Lable for X")
plt.ylabel("Lable for Y")
plt.show()
```

# Use of styles in Matplotlib 

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('seaborn')
x = [1,5,8]
y = [34,50,48]
x2 = [3,5,7]
y2 = [27,44,48]

plt.plot(x,y, label="first", linewidth=3)
plt.plot(x2,y2, label="second", linewidth=3)
plt.legend()
plt.show()
```
